```python
import string
import matplotlib.pyplot as plt
from collections import Counter, defaultdict
import nltk
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.util import ngrams
import re


paragraph = "Food, an essential aspect of human existence, transcends mere sustenance to become a celebration of culture, taste, and tradition. Special characters like *, &, and punctuation marks such as exclamation points and question marks adorn the culinary landscape, adding flavor to our gastronomic experiences. Amidst the myriad flavors and textures, stop words like 'the' 'to' and 'and' effortlessly weave through the tapestry of our culinary conversations, connecting us to the rich heritage of culinary delights. While some may advocate for indulgence, others may advise restraint; however, don’t let limitations stifle culinary exploration. In my email address, sowmya@gmail.com and phone number +91 9876543210 facilitate culinary exploration, connecting food enthusiasts to culinary experts and gourmet experiences. As we savor each bite, significant dates like 27-March-2000 serve as reminders of shared meals and cherished memories, nourishing both body and soul. So, let us embrace the culinary journey with gusto, for in food, we find not just sustenance, but also joy, comfort, and connection."
```

```python
# Tokenization
tokens = paragraph.split()
# Count unique tokens
unique_tokens = set(tokens)
num_unique_tokens = len(unique_tokens)
print("Number of unique tokens:", num_unique_tokens)
```

This can be valuable for text analysis tasks such as authorship attribution, sentiment analysis, or topic modeling. For example, in authorship attribution, comparing the vocabulary richness of different authors can help identify their distinctive writing styles.

```python
import string

# Count punctuation characters
num_punctuations = sum([1 for char in paragraph if char in string.punctuation])
print("Number of punctuation characters:", num_punctuations)

# Remove punctuations
cleaned_paragraph = paragraph.translate(str.maketrans('', '', string.punctuation))
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import nltk
from nltk.corpus import stopwords

# Download stopwords
nltk.download('stopwords')

# Tokenize and count stopwords
stop_words = set(stopwords.words('english'))
stopwords_count = Counter(word for word in tokens if word.lower() in stop_words)

# Plot
plt.bar(stopwords_count.keys(), stopwords_count.values())
plt.xlabel('Stopwords')
plt.ylabel('Frequency')
plt.title('Distribution of Stopwords')
plt.xticks(rotation=45)
plt.show()
```

c. Plotting Distribution of Stopwords: Examining the distribution of stopwords (common words like "and", "the", "is") can offer insights into the structure and readability of the text. Understanding the frequency of stopwords can aid in text preprocessing tasks such as stopword removal, which can improve the performance of downstream NLP tasks like information retrieval or text classification.

```python
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
filtered_paragraph = ' '.join(filtered_tokens)
```

```python
# Check if filtered_tokens contains any words
print("Filtered tokens:", filtered_tokens)

# Check if filtered_paragraph is not empty
print("Filtered paragraph:", filtered_paragraph)

# Print length of filtered_tokens
print("Number of filtered tokens:", len(filtered_tokens))
```

d. Removing Stopwords: Removing stopwords from text is a crucial preprocessing step in many NLP tasks. By eliminating noise words, we focus on content-bearing words, leading to more accurate analysis and interpretation. This process is essential for tasks like sentiment analysis, where stopwords may not contribute to the sentiment polarity of the text.

```python
from nltk import pos_tag
from collections import defaultdict

# POS tagging
pos_tags = pos_tag(tokens)
pos_counts = defaultdict(int)

# Count POS tags
for word, tag in pos_tags:
    pos_counts[tag] += 1

# Plot
plt.bar(pos_counts.keys(), pos_counts.values())
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.title('Distribution of POS Tags')
plt.xticks(rotation=45)
plt.show()
```

e. Plotting Distribution of POS Tags: Analyzing the distribution of parts-of-speech (POS) tags provides insights into the grammatical structure and semantic meaning of the text. This information is valuable for tasks such as named entity recognition, syntactic parsing, or machine translation, where understanding the grammatical roles of words is crucial for accurate analysis and translation.

```python
from nltk.stem import WordNetLemmatizer

# Download WordNet
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
unique_lemmas = set(lemmatizer.lemmatize(word) for word in tokens)
num_unique_lemmas = len(unique_lemmas)
print("Number of unique lemmas:", num_unique_lemmas)
```

f. Counting Unique Lemmas: Counting unique lemmas (base forms of words) helps in reducing the dimensionality of the vocabulary while preserving semantic information. This is beneficial for tasks such as text classification, where using lemmatized words can improve model generalization and reduce overfitting by collapsing inflected forms of words into their base forms.

```python
word_counts = Counter(filtered_tokens)
top_10_words = word_counts.most_common(10)

plt.bar([pair[0] for pair in top_10_words], [pair[1] for pair in top_10_words])
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Most Occurring Words')
plt.xticks(rotation=45)
plt.show()
```

g. Plotting Frequency Distribution of Words: Visualizing the frequency distribution of words reveals important insights into the content and key themes of the text. This information can be used for tasks such as keyword extraction, content summarization, or topic modeling, where identifying frequently occurring words helps in identifying important concepts and topics within the text.

```python
from nltk.util import ngrams

# Generate ngrams
bigrams = list(ngrams(filtered_tokens, 2))
trigrams = list(ngrams(filtered_tokens, 3))
quadgrams = list(ngrams(filtered_tokens, 4))

num_unique_bigrams = len(set(bigrams))
num_unique_trigrams = len(set(trigrams))
num_unique_quadgrams = len(set(quadgrams))

print("Number of unique bigrams:", num_unique_bigrams)
print("Number of unique trigrams:", num_unique_trigrams)
print("Number of unique quadgrams:", num_unique_quadgrams)
```

h. Finding Unique Bigrams, Trigrams, and Quadgrams: Identifying unique n-grams (sequences of n words) provides insights into collocations and patterns of co-occurring words in the text. This is valuable for tasks such as phrase extraction, where identifying meaningful multi-word expressions helps in capturing nuanced semantic relationships between words.

```python
import re
import datetime
# Corrected regex pattern for date
date_pattern = re.compile(r'\b(?:0?[1-9]|[12][0-9]|3[01])-(?:January|February|March|April|May|June|July|August|September|October|November|December)-\d{4}\b')

# Find and convert dates
dates = date_pattern.findall(paragraph)
converted_dates = [datetime.datetime.strptime(date, '%d-%B-%Y').strftime('%d-%m-%Y') for date in dates]

print("Extracted dates:", dates)
print("Converted dates:", converted_dates)
```

i. Finding and Formatting Dates: Extracting and formatting dates in text enables temporal analysis and tracking of events over time. This information is useful for tasks such as trend analysis, event detection, or timeline generation, where understanding temporal patterns is crucial for decision-making and analysis.

```python
# Extract years from converted dates
years = [date.split('-')[2] for date in converted_dates]

# Count frequency of each year
year_counts = Counter(years)

# Plot distribution of years
plt.bar(year_counts.keys(), year_counts.values())
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.title('Distribution of Years')
plt.xticks(rotation=45)
plt.show()
```

```python
import re
from collections import Counter
import matplotlib.pyplot as plt

# Sample paragraph containing years
paragraph = "The project started in 2019 and ended in 2021. We also had significant progress in 2020."

# Define a regex pattern to match years
year_pattern = re.compile(r'\b\d{4}\b')

# Find all years in the text
years = year_pattern.findall(paragraph)

# Count the frequency of each year
year_counts = Counter(years)

# Plot distribution of years
plt.figure(figsize=(10, 6))  # Set figure size
plt.bar(year_counts.keys(), year_counts.values(), color='skyblue')  # Bar plot
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.title('Distribution of Years')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
import re

# Define a regex pattern to match phone numbers
phone_pattern = re.compile(r'\b(?:\+\d{1,3}[-\.\s]?)?\(?\d{3}\)?[-\.\s]?\d{3}[-\.\s]?\d{4}\b')

# Find all phone numbers in the text
phone_numbers = phone_pattern.findall(paragraph)

# Validate each phone number
valid_phone_numbers = []
for number in phone_numbers:
    if re.match(r'^(\+\d{1,3}[-\.\s]?)?\(?\d{3}\)?[-\.\s]?\d{3}[-\.\s]?\d{4}$', number):
        valid_phone_numbers.append(number)

# Remove invalid phone numbers from the text
cleaned_paragraph = re.sub(phone_pattern, '', paragraph)

# Print extracted and valid phone numbers
print("Extracted phone numbers:", phone_numbers)
print("Valid phone numbers:", valid_phone_numbers)
print("Cleaned paragraph:", cleaned_paragraph)
```

k. Identifying and Validating Phone Numbers: Extracting and validating phone numbers in text enables contact information extraction and data cleaning. This is useful for tasks such as customer relationship management, data validation, or text preprocessing, where ensuring the accuracy and integrity of contact information is crucial for effective communication and analysis.
