```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

corpus = """
Transformer is an exceptional innovation in the field of Deep Learning, contributed by Ashish Vaswani et al. (2017), Google.
The transformer is the most influential Neural Network model that has shown outstanding performance on various NLP tasks including Machine Reading Comprehension,
Machine translation, and sentence classification. Attention mechanism and parallelization are the prominent features in the transformers.
Consequently, it can facilitate long-range dependencies without any gradient vanishing or gradient explosion problems and it overcomes the drawbacks of the existing methods such as RNN and LSTM.
The transformer is executed with an encoder-decoder mechanism and the original article of transformers # “Attention All You Need”.
"""
```

```python
# Tokenization
words = word_tokenize(corpus)
sentences = sent_tokenize(corpus)
print("Tokenized words:", words)
print("Tokenized sentences:", sentences)
```

```python
# Stopwords Removal
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.lower() not in stop_words]
print("Words after stopwords removal:", filtered_words)
```

```python
# Punctuation Removal
tokenizer = RegexpTokenizer(r'\w+')
filtered_words_no_punct = tokenizer.tokenize(corpus)
print("Words after punctuation removal:", filtered_words_no_punct)
```

```python
# Frequency Distribution
fdist = FreqDist(filtered_words_no_punct)
fdist.plot()
plt.show()
```

```python
dist = FreqDist(filtered_words)
print(dist)
dist.plot()
plt.show()
```

```python
# Punctuation Removal
tokenizer = RegexpTokenizer(r'\w+')
filtered_words_no_punct = tokenizer.tokenize(' '.join(filtered_words))

# Frequency Distribution after Stopwords and Punctuation removal
fdist = FreqDist(filtered_words_no_punct)

# Plotting the frequency distribution
plt.figure(figsize=(12, 6))
fdist.plot(30, cumulative=False)
plt.show()
```

```python
# Stemming and Lemmatization
porter = PorterStemmer()
snowball = SnowballStemmer(language='english')
stemmed_words_porter = [porter.stem(word) for word in filtered_words]
stemmed_words_snowball = [snowball.stem(word) for word in filtered_words]
print("Words after stemming (Porter):", stemmed_words_porter)
print("Words after stemming (Snowball):", stemmed_words_snowball)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
print("Words after lemmatization:", lemmatized_words)
```

```python
# Part-of-Speech Tagging
pos_tags = nltk.pos_tag(filtered_words)
print("Part-of-Speech tags:", pos_tags)
```

```python

from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


text = "Mickey Mouse, a cheerful and optimistic mouse clad in red shorts and yellow shoes, is the iconic mascot of The Walt Disney Company. Debuting in 1928, this spunky character has charmed audiences for generations with his adventures and can-do attitude."
```

```python
words = text.split()
```

```python
skipgram_model = Word2Vec(sentences=[words], vector_size=100, window=5, min_count=1, sg=1)
cbow_model = Word2Vec(sentences=[words], vector_size=100, window=5, min_count=1, sg=0)
```

```python
skipgram_vocabulary = skipgram_model.wv.index_to_key
skipgram_word_vectors = skipgram_model.wv[skipgram_vocabulary]
```

```python
cbow_vocabulary = cbow_model.wv.index_to_key
cbow_word_vectors = cbow_model.wv[cbow_vocabulary]
```

```python
pca = PCA(n_components=2)
```

```python
skipgram_word_vectors_2D = pca.fit_transform(skipgram_word_vectors)
```

```python
cbow_word_vectors_2D = pca.fit_transform(cbow_word_vectors)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(skipgram_word_vectors_2D[:, 0], skipgram_word_vectors_2D[:, 1])
for i, word in enumerate(skipgram_vocabulary):
    plt.annotate(word, xy=(skipgram_word_vectors_2D[i, 0], skipgram_word_vectors_2D[i, 1]), ha='right')
plt.title('Skip-gram Word Vectors')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(cbow_word_vectors_2D[:, 0], cbow_word_vectors_2D[:, 1])
for i, word in enumerate(cbow_vocabulary):
    plt.annotate(word, xy=(cbow_word_vectors_2D[i, 0], cbow_word_vectors_2D[i, 1]), ha='right')
plt.title('CBOW Word Vectors')
plt.show()
```

Skip-gram Model:

 Skip-gram is a Word2Vec model that predicts context words given a target word.
It trains a neural network to predict surrounding words based on a target word.
Captures semantic relationships between words and represents them as dense vectors.
 for large corpora words.
Continuous Bag of Words (CBOW) Model:

CBOW Word2Vec model predicts the target word based on context words.
 Learns by predicting the target word using surrounding context words within a window. Efficient for smaller datasets, quick training times, and tasks like predicting missing words.
